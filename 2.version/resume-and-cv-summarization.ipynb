We will first load a black SpaCy english model. Then we will write a function which will take the training data as the input. In the function, first we will add a ner i.e. Named Entity Recognition in the last position in the pipeline. Then we will add our custom labels in the pipeline.

```python
# Loading the blank SpaCy english model
nlp = spacy.load('en_core_web_sm')
```

```python
# Check 'ner' component exits ?
if 'ner' not in nlp.pipe_names:
    ner = nlp.create_pipe('ner')
    nlp.add_pipe(ner, last = True)
else:
    ner = nlp.get_pipe("ner")  # Get the NER component
```

```python
nlp.pipe_names
```

```python
# Adding Labels
for text, annotations in train_data:
    print(annotations)
    for ent in annotation['entities']:
        ner.add_label(ent[2])
```

```python
# Disabling Other Pipes
other_pipes = [pipe for pipe in nlp.pipe_names if pipe != 'ner']
```

```python
with nlp.disable_pipes(*other_pipes):  # only train NER
    optimizer = nlp.begin_training()
    for itn in range(500):
        print("Starting iteration " + str(itn))
        random.shuffle(train_data)
        losses = {}
        index = 0
        for text, annotations in train_data:
            try:
                nlp.update(
                    [text],         # batch of texts
                    [annotations],  # batch of annotations
                    drop=0.3,       # dropout - make it harder to memorise data
                    sgd=optimizer,  # callable to update weights
                    losses=losses)
            except Exception as e:
                pass
                
        print(losses)
```

```python
# Let's save the model for further use
nlp.to_disk('nlp_model.h5')
```
